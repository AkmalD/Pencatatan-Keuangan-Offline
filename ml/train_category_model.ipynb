# Training Model Kategori Transaksi

Notebook ini memakai kode yang sama dengan `train_category_model.py`, lalu mengekspor model statis ke asset Android.

```python
from pathlib import Path
import sys

ROOT = Path.cwd()
if (ROOT / 'ml').exists():
    sys.path.insert(0, str(ROOT / 'ml'))
    CSV_PATH = ROOT / 'ml' / 'training_data.csv'
    OUTPUT_PATH = ROOT / 'app' / 'src' / 'main' / 'assets' / 'category_model.json'
else:
    sys.path.insert(0, str(ROOT))
    CSV_PATH = ROOT / 'training_data.csv'
    OUTPUT_PATH = ROOT.parent / 'app' / 'src' / 'main' / 'assets' / 'category_model.json'

from train_category_model import build_model, evaluate, export_model, load_rows, predict
```

```python
rows = load_rows(CSV_PATH)
model = build_model(rows, alpha=1.0)
accuracy = evaluate(model, rows)

print(f'Training rows: {model["metadata"]["training_rows"]}')
print(f'Vocabulary size: {model["metadata"]["vocabulary_size"]}')
print(f'Training accuracy: {accuracy:.2%}')
```

```python
examples = [
    'investasi 100 rb di bibit',
    'beli makan 10 ribu di warteg',
    'gaji bulanan masuk 5 juta',
    'bayar listrik 200 ribu',
]

for example in examples:
    category, confidence = predict(model, example)
    print(f'{example} -> {category} ({confidence:.1%})')
```

```python
export_model(model, OUTPUT_PATH)
print(f'Model exported to: {OUTPUT_PATH}')
```
